```python
from pathlib import Path
import pandas as pd
from matplotlib import pyplot
import collections
import numpy as np
import random
chemin = "C:/Users/Ignifuge/Documents/GitHub/BC/data_cog"
chemin_adulte = chemin + "/Adult"
chemin_enfant = chemin + "/Child"
```

```python
list_paths = list(Path(chemin_adulte).glob('**/*')) + list(Path(chemin_enfant).glob('**/*'))
```

```python
def get_length(path):
    for chemin0 in list(Path(chemin + "/" + path).glob('**/*.csv')):
        list_objets = pd.read_csv(chemin0, names=["Tier", "Sender", "Creation Date", "Start Time", "Modification Date", "End Time", "_1", "_2", "Type", "Comment"])
        return max(list_objets["End Time"])
```

```python
def get_res(path, delta):
    
    list_objets = pd.read_csv(path, names=["Tier", "Sender", "Creation Date", "Start Time", "Modification Date", "End Time", "_1", "_2", "Type", "Comment"])
    feature_list = {}
    for objet in list_objets.iterrows():
        if str(objet[1]["Comment"]) == "nan":
            action = str(objet[1]["Type"])
        else:
            action = str(objet[1]["Type"]) + ", " + str(objet[1]["Comment"])
        add = (float(objet[1]["Start Time"]), float(objet[1]["End Time"]), action)
        if objet[1]["Type"] != "Speech":
            
            if action not in feature_list.keys():
                feature_list[action] = [[], [], None]
                
            if objet[1]["Sender"] in ["Adult1", "Parent"]:
                feature_list[action][0].append(add)
            else:
                feature_list[action][1].append(add)
    ret = {}
    for feature in feature_list:
        ret[feature] = count_mimicry(feature_list[feature][0], feature_list[feature][1], delta)
        
    return ret
```

```python
# high levels
def count_mimicry(lstA, lstB, delta_t=0):
    """Count the occurences of B mimicking A by delta_t.
   
    This implementation counts mimicry based on method in [1]
    and also returns the instances of mimickry.

    The times in each of lstA and lstB cannot overlap internally.
    They have to be successive segments in each of lstA and lstB.

    [1] Feese, Sebastian, et al. "Quantifying behavioral mimicry by automatic 
    detection of nonverbal cues from body motion." 2012 International Conference 
    on Privacy, Security, Risk and Trust and 2012 International Conference on 
    Social Computing. IEEE, 2012.
    
    Args:
        lstA (list): list of tuples (start, stop, label) of expressions mimicked.
        lstB (list): list of tuples (start, stop, label) of expressions mimicking.
        delta_t (int, optional): Defaults to 0.
                                Time after which expression occuring still counts as mimicry.
                                Should be in the same unit as the times in lstA and lstB.    
    Returns:
        int: number of times B mimicked A (=len(the list described below)).
        list: [(indA, indB),...]
              where the indB element of B mimick the indA element of A
              following the definition of mimickry described in the reference above.
    """

    indA = 0
    indB = 0
    count = 0  # number of mimicry events
    mimic_ind = []  # indices of mimicry events in lstB
    if len(lstA) == 0 or len(lstB) == 0:  # if at least one of them has no expression
        return 0, []
    while indA < len(lstA) and indB < len(lstB):
        if lstB[indB][0] <= lstA[indA][0]:
            indB += 1
        elif (
            lstB[indB][0] > lstA[indA][0] and (lstB[indB][0] - delta_t) <= lstA[indA][1]
        ):
            # avoid double counting incase delta_t is > (lstA[indA+1][0] - lstA[indA][1])
            if (indA + 1) < len(lstA):
                if lstB[indB][0] > lstA[indA + 1][0]:
                    indA += 1  # skip to next lstA expression
                    continue
            count += 1
            mimic_ind.append((indA, indB))
            # if no double counting
            # check if several expressions from B overlap with A's
            while lstB[indB][1] <= lstA[indA][1]:
                indB += 1  # skip to the following expression untill no more overlapping
                if indB == len(lstB):
                    break
            indA += 1
        elif (lstB[indB][0] - delta_t) > lstA[indA][1]:
            indA += 1
    return count, mimic_ind
```

```python
def get_feature_par_fichier(delta=10):
    feature_par_fichier = {}
    for path in list_paths:
        if ".DS_Store" not in str(path):
            # A moins que le dossier ne contienne pas de .csv
            if len(list(Path(path).glob('**/*.csv'))) != 0:
                # On récupère l'ense
                feature_par_fichier[str(path).replace(str(Path(chemin)) + "\\", "")] = get_res(list(Path(path).glob('**/*.csv'))[0], delta)
    return feature_par_fichier
```

```python
get_feature_par_fichier(10)
```

```python
def afficher_graphe(relation):
    feature_gen = {}
    delta = 10
    feature_par_fichier = get_feature_par_fichier(delta)
    for fichier in feature_par_fichier:
        if fichier.startswith(relation):
            for feature in feature_par_fichier[fichier]:
                if feature not in feature_gen:
                    feature_gen[feature] = 0
                feature_gen[feature] += feature_par_fichier[fichier][feature][0]
    somme = 0
    for feature in feature_gen:
        somme += feature_gen[feature]
    for feature in feature_gen:
        feature_gen[feature] = feature_gen[feature]/somme
    pyplot.figure(figsize = (12, 12))
    pyplot.title("Fréquence (en %) d'occurence de features classées par type de conversation (Adult/" + relation + "), delta=" + str(delta))
    values = sorted(list(feature_gen.values()))
    keys = sorted(list(feature_gen.keys()), key=lambda x: feature_gen[x])
    labels = []
    for i, label in enumerate(keys):
        labels.append(label + " (" + str(feature_gen[label]*somme) + ")")
    pyplot.pie(values, labels = labels, normalize = True)
```

```python
afficher_graphe("Adult")
```

```python
afficher_graphe("Child")
```

```python
def afficher_graphe2(relation):
    feature_gen = {}
    delta_ens = range(5, 20, 2)
    for delta in delta_ens:
        feature_par_fichier = get_feature_par_fichier(delta)
        for fichier in feature_par_fichier:
            if fichier.startswith(relation):
                for feature in feature_par_fichier[fichier]:
                    if not feature.startswith("LS"):
                        if feature not in feature_gen:
                            feature_gen[feature] = {}
                        if delta not in feature_gen[feature]:
                            feature_gen[feature][delta] = 0
                        feature_gen[feature][delta] += feature_par_fichier[fichier][feature][0]
    pyplot.figure(figsize = (12, 12))
    pyplot.title("Effectifs des occurences de features classées par type de conversation (Adult/" + relation + ")")
    known_y = []
    for feature in feature_gen:
        values = list(feature_gen[feature].values())
        keys = list(feature_gen[feature].keys())
        pyplot.plot(keys, values, label=feature)
        pyplot.xlabel("Delta (en secondes)")
        pyplot.ylabel("Nombre de fois que la feature apparaît")
        n = 1
        while (keys[-n], values[-n]) in known_y and n < len(keys):
            n += 1
        pyplot.text(keys[-n] + 0.2, values[-n] - ((n + 1) % 2)/10, feature)
        known_y.append((keys[-n], values[-n]))
```

```python
afficher_graphe2("Adult")
```

```python
afficher_graphe2("Child")
```

```python
def process2(keys, values):
    indices = sorted(list(range(len(values))), key=lambda i: keys[i])
    keys_ret = [keys[i] for i in indices]
    values_ret = [values[i] for i in indices]
    return keys_ret, values_ret
    
def afficher_graphe3(relation):
    feature_gen1 = {}
    feature_gen2 = {}
    delta = 20
    feature_par_fichier = get_feature_par_fichier(delta)
    for fichier in feature_par_fichier:
         if fichier.startswith(relation):
            for feature in feature_par_fichier[fichier]:
                if not feature.startswith("LS"):
                        if feature not in feature_gen1:
                            feature_gen1[feature] = 0
                        if feature not in feature_gen2:
                            feature_gen2[feature] = 0
                else:
                    continue
                for occurence in feature_par_fichier[fichier][feature][1]:
                    feature_gen = None
                    if occurence[0] < occurence[1]:
                        feature_gen = feature_gen1
                    else:
                        feature_gen = feature_gen2      
                    feature_gen[feature] += 1 / get_length(fichier)
                if feature_gen1[feature] == 0 and feature_gen2[feature] == 0:
                    del feature_gen1[feature]
                    del feature_gen2[feature]
    pyplot.figure(figsize = (12, 12))
    pyplot.title("Fréquence des occurences des features classées par type de conversation (Adult/" + relation + ")")
    keys1 = list(feature_gen1.keys())
    keys2 = list(feature_gen2.keys())
    values1 = list(feature_gen1.values())
    values2 = list(feature_gen2.values())
    keys1, values1 = process(keys1, values1)
    keys2, values2 = process(keys2, values2)
    x = np.arange(len(keys1))
    pyplot.bar(x-0.2, values1, 0.2, label=feature, color=["red"])
    pyplot.bar(x+0.2, values2, 0.2, label=feature, color=["green"])
    pyplot.xticks(x, keys1)
    pyplot.xlabel("Delta (en secondes)")
    pyplot.ylabel("Effectif du mimétisme/Longueur de la conversation")
    
    def get_legend(relation):
        i1, i2 = "", ""
        if relation == "Adult":
            i1 = relation + "1"
            i2 = relation + "2"
        else:
            i1 = "Adult"
            i2 = relation
        return [i1 + " then " + i2, i2 + " then " + i1]
    
    pyplot.legend(get_legend(relation))
```
