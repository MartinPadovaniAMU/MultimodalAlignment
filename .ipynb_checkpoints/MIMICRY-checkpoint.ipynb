```python
from pathlib import Path
import pandas as pd
from matplotlib import pyplot
import collections
import numpy as np
import random
import json
from runpy import run_path
from os.path import join

# Load local paths and variables
data = None
with open("config.json", "r") as read_file:
    data = json.load(read_file)

output_path = join(data["general"]["output_path"], "mimicry")
path = join(data["mimicry"]["data_repository_path"], "data_cog")
adult_path = join(path, "Adult")
child_path = join(path, "Child")
column_names = ["Tier", "Sender", "Creation Date", "Start Time", "Modification Date", "End Time", "_1", "_2", "Type", "Comment"]
```

```python
# Get the complete list of csv directories
list_paths = list(Path(adult_path).glob('**/*')) + list(Path(child_path).glob('**/*'))
```

```python
def get_length(subpath):
    """
    Returns the total length of the conversion (i.e. the end time of the last utterance/the maximal end time) for a given subpath "(Adult|Child)\\Filecode"
    """
    for csv_path in list(Path(path + "/" + subpath).glob('**/*.csv')):
        list_objects = pd.read_csv(csv_path, names=column_names)
        return max(list_objects["End Time"])
```

```python
def get_res(path, delta):
    """
    get_res (get result) numbers the mimicry events for each feature for a given file path and a given delta
    
    path : complete path of the .csv file from which it will extract the data
    delta : -> see 'count_mimicry' below
    
    Returns : a Python dictionary with the following format : {"feature1": count1, "feature2": count2, ...}
    """
    list_objets = pd.read_csv(path, names=column_names)
    
    # A temporary object {..., "feature": [listA, listB], ...} containing two out of three arguments for the count_mimicry 
    # function (see count_mimicry for informations on listA and listB)
    feature_list = {}
    
    # For each row in the csv file
    for objet in list_objets.iterrows():
        
        # Concatenate comments to the type of feature if they exist
        if str(objet[1]["Comment"]) == "nan":
            action = str(objet[1]["Type"])
        else:
            action = str(objet[1]["Type"]) + ", " + str(objet[1]["Comment"])
        
        # This row will be appended to the corresponding list of events 
        row = (float(objet[1]["Start Time"]), float(objet[1]["End Time"]), action)
        
        # If the feature is that of a non-verbal expression
        if objet[1]["Type"] != "Speech":
            
            # Create the two lists listA and listB
            if action not in feature_list.keys():
                feature_list[action] = [[], []]
            
            # If the sender is the interviewer
            if objet[1]["Sender"] in ["Adult1", "Parent"]:
                feature_list[action][0].append(row)
            # Else if the sender is the interviewee
            else:
                feature_list[action][1].append(row)
    
    # Final object to be returned
    ret = {}
    for feature in feature_list:
        ret[feature] = count_mimicry(feature_list[feature][0], feature_list[feature][1], delta)
        
    return ret
```

```python
# Imported from the IBPY repository (https://github.com/kelhad00/IBPY/blob/master/interaction_analysis.py)

def count_mimicry(lstA, lstB, delta_t=0):
    """Count the occurences of B mimicking A by delta_t.
   
    This implementation counts mimicry based on method in [1]
    and also returns the instances of mimickry.

    The times in each of lstA and lstB cannot overlap internally.
    They have to be successive segments in each of lstA and lstB.

    [1] Feese, Sebastian, et al. "Quantifying behavioral mimicry by automatic 
    detection of nonverbal cues from body motion." 2012 International Conference 
    on Privacy, Security, Risk and Trust and 2012 International Conference on 
    Social Computing. IEEE, 2012.
    
    Args:
        lstA (list): list of tuples (start, stop, label) of expressions mimicked.
        lstB (list): list of tuples (start, stop, label) of expressions mimicking.
        delta_t (int, optional): Defaults to 0.
                                Time after which expression occuring still counts as mimicry.
                                Should be in the same unit as the times in lstA and lstB.    
    Returns:
        int: number of times B mimicked A (=len(the list described below)).
        list: [(indA, indB),...]
              where the indB element of B mimick the indA element of A
              following the definition of mimickry described in the reference above.
    """

    indA = 0
    indB = 0
    count = 0  # number of mimicry events
    mimic_ind = []  # indices of mimicry events in lstB
    if len(lstA) == 0 or len(lstB) == 0:  # if at least one of them has no expression
        return 0, []
    while indA < len(lstA) and indB < len(lstB):
        if lstB[indB][0] <= lstA[indA][0]:
            indB += 1
        elif (
            lstB[indB][0] > lstA[indA][0] and (lstB[indB][0] - delta_t) <= lstA[indA][1]
        ):
            # avoid double counting incase delta_t is > (lstA[indA+1][0] - lstA[indA][1])
            if (indA + 1) < len(lstA):
                if lstB[indB][0] > lstA[indA + 1][0]:
                    indA += 1  # skip to next lstA expression
                    continue
            count += 1
            mimic_ind.append((indA, indB))
            # if no double counting
            # check if several expressions from B overlap with A's
            while lstB[indB][1] <= lstA[indA][1]:
                indB += 1  # skip to the following expression untill no more overlapping
                if indB == len(lstB):
                    break
            indA += 1
        elif (lstB[indB][0] - delta_t) > lstA[indA][1]:
            indA += 1
    return count, mimic_ind
```

```python
def get_total_feature_inventory(delta=10):
    total_feature_inventory = {}
    
    # For each filecode directory
    for subpath in list_paths:
        if ".DS_Store" not in str(subpath):
            
            # Unless there is no .csv in the selected path
            if len(list(Path(subpath).glob('**/*.csv'))) != 0:
                
                # We load the inventory of features for the current conversation
                total_feature_inventory[str(subpath).replace(str(Path(path)) + "\\", "")] = get_res(list(Path(subpath).glob('**/*.csv'))[0], delta)
    
    return total_feature_inventory
```

```python
def show_graph(conversation_type):
    feature_glob = {}
    delta = 10
    # Retrieve the feature inventory indexed by (conversation_type (Adult/Child) + "\\" + filecode)
    total_feature_inventory = get_total_feature_inventory(delta)
    
    # For each file
    for file in total_feature_inventory:
        
        # We select files with the right conversation type
        if file.startswith(conversation_type):
            for feature in total_feature_inventory[file]:
                
                # We create the key if it doesn't already exist
                if feature not in feature_glob:
                    feature_glob[feature] = 0
                
                # We add the file dependent count to the global count
                feature_glob[feature] += total_feature_inventory[file][feature][0]
    
    # We number the total of features
    total_of_events = 0
    for feature in feature_glob:
        total_of_events += feature_glob[feature]
    
    # Then, we divide each count by that value
    for feature in feature_glob:
        feature_glob[feature] = feature_glob[feature]/total_of_events
    
    # We plot the result using a pie chart and save it
    fig = pyplot.figure(figsize = (12, 12))
    fig.patch.set_facecolor('white') # Without this instruction, the whole background stays transparent, even if the plot itself isn't
    pyplot.title("Fréquence (en %) d'occurence de features classées par type de conversation (Adult/" + conversation_type + "), delta=" + str(delta))
    # The values are sorted by descending
    values = sorted(list(feature_glob.values()))
    keys = sorted(list(feature_glob.keys()), key=lambda x: feature_glob[x])
    labels = []
    for i, label in enumerate(keys):
        labels.append(label + " (" + str(feature_glob[label]*total_of_events) + ")")
    pyplot.pie(values, labels = labels, normalize = True)
    pyplot.savefig(output_path + "graph1" + conversation_type + ".png")
```

```python
show_graph("Adult")
```

```python
show_graph("Child")
```

```python
def show_graph2(conversation_type):
    feature_glob = {}
    delta_ens = range(5, 20, 2)

    # For each delta (this time, the global count depends on delta)
    for delta in delta_ens:
        total_feature_inventory = get_total_feature_inventory(delta)
        
        # For each processed file
        for fichier in total_feature_inventory:
            if fichier.startswith(conversation_type):
                
                # For each feature inventoried in that file
                for feature in total_feature_inventory[fichier]:
                    
                    # We then decided to remove LS (Looking at the screen) because it doesn't carry any information
                    if not feature.startswith("LS"):
                        
                        # We create the keys if they don't already exist
                        if feature not in feature_glob:
                            feature_glob[feature] = {}
                            
                        if delta not in feature_glob[feature]:
                            feature_glob[feature][delta] = 0
                        
                        # We add the file dependent count to the global count
                        feature_glob[feature][delta] += total_feature_inventory[fichier][feature][0]
                        
    # We plot the result using a plot with multiple curves and save it
    fig = pyplot.figure(figsize = (12, 12))
    fig.patch.set_facecolor('white') # Without this instruction, the whole background stays transparent, even if the plot itself isn't
    pyplot.title("Effectifs des occurences de features classées par type de conversation (Adult/" + conversation_type + ")")
    known_y = []
    for feature in feature_glob:
        values = list(feature_glob[feature].values())
        keys = list(feature_glob[feature].keys())
        pyplot.plot(keys, values, label=feature) # x = delta (keys of feature_glob[feature] for each feature)
        pyplot.xlabel("Delta (en secondes)")
        pyplot.ylabel("Nombre de fois que la feature apparaît")
        
        # The purpose of this code is to insure that the feature labels are correctly placed
        n = 1
        while (keys[-n], values[-n]) in known_y and n < len(keys):
            n += 1
        pyplot.text(keys[-n] + 0.2, values[-n] - ((n + 1) % 2)/10, feature)
        known_y.append((keys[-n], values[-n]))
    pyplot.savefig(output_path + "graph2" + conversation_type + ".png")
```

```python
show_graph2("Adult")
```

```python
show_graph2("Child")
```

```python
def process(keys, values):
    """
    A very short function to sort both keys and values by key.
    """
    indices = sorted(list(range(len(values))), key=lambda i: keys[i])
    keys_ret = [keys[i] for i in indices]
    values_ret = [values[i] for i in indices]
    return keys_ret, values_ret

def get_legend(conversation_type):
    """
    Generate a legend depending on the type of conversation
    """
    i1, i2 = "", ""
    if conversation_type == "Adult":
        i1 = conversation_type + "1"
        i2 = conversation_type + "2"
    else:
        i1 = "Adult"
        i2 = conversation_type
    return [i1 + " then " + i2, i2 + " then " + i1]
    
def show_graph3(conversation_type):
    # Two global inventories because this time, we try to determine whether the interviewer or the interviewee initiated the event
    feature_glob1 = {}
    feature_glob2 = {}
    
    delta = 10
    total_feature_inventory = get_total_feature_inventory(delta)
    # For each processed file
    for fichier in total_feature_inventory:
         if fichier.startswith(conversation_type):
                
            # For each feature inventoried in that file
            for feature in total_feature_inventory[fichier]:
                
                # We still exclude LS as a part of the graph and initialize the other counts
                if not feature.startswith("LS"):
                        if feature not in feature_glob1:
                            feature_glob1[feature] = 0
                        if feature not in feature_glob2:
                            feature_glob2[feature] = 0
                else:
                    continue
                
                # We go through the list of events for a given feature
                for occurence in total_feature_inventory[fichier][feature][1]:
                    
                    # We select the right inventory before filling it
                    feature_glob = None
                    if occurence[0] < occurence[1]:
                        feature_glob = feature_glob1
                    else:
                        feature_glob = feature_glob2 
                    
                    # We divide the difference between the timestamp of the first and the second utterance and divide it by the total length of the file
                    feature_glob[feature] += abs(occurence[1] - occurence[0]) / get_length(fichier)
                
                # If, in the end, the two inventories are still empty for this feature, we can remove it from the plot
                if feature_glob1[feature] == 0 and feature_glob2[feature] == 0:
                    del feature_glob1[feature]
                    del feature_glob2[feature]
                    
    fig = pyplot.figure(figsize = (12, 12))
    fig.patch.set_facecolor('white')
    pyplot.title("Fréquence des occurences des features classées par type de conversation (Adult/" + conversation_type + ")")
    
    # Dividing and sorting the dictionaries in order to plot them
    keys1 = list(feature_glob1.keys())
    keys2 = list(feature_glob2.keys())
    values1 = list(feature_glob1.values())
    values2 = list(feature_glob2.values())
    keys1, values1 = process(keys1, values1)
    keys2, values2 = process(keys2, values2)
    x = np.arange(len(keys1))
    
    # We group each feature count by pairs, where the first bar numbers "participant 1 then participant 2" events 
    # and the second bar numbers "participant 2 then participant 1" events 
    pyplot.bar(x-0.2, values1, 0.2, color=["red"])
    pyplot.bar(x+0.2, values2, 0.2, color=["green"])
    pyplot.xticks(x, keys1)
    pyplot.xlabel("Delta (en secondes)")
    pyplot.ylabel("Effectif du mimétisme/Longueur de la conversation")
    
    pyplot.legend(get_legend(conversation_type))
    
    pyplot.savefig(output_path + "graph3" + conversation_type + ".png")
```

```python
show_graph3("Adult")
```

```python
show_graph3("Child")
```
